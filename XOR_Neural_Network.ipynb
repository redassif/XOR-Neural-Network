```python
#Reda Assif


import numpy as np
import random
import math
import matplotlib.pyplot as plt

q = int(input("Enter 1 for batch training and 2 for stochastic training"))
if(q == 1):
  inputs = np.array([[0, 0], [1, 0], [0, 1], [1, 1]])
  outputs = np.array([0,1,1,0], dtype = float)
  L = 0.1
  k = len(inputs)
  f = open("NNModelParameters.txt", "w")


  weights = np.random.randn(6,1)
  weights[0] = random.uniform(0,1)
  weights[1] = random.uniform(0,1)
  weights[2] = random.uniform(0,1)
  weights[3] = random.uniform(0,1)
  weights[4] = random.uniform(0,1)
  weights[5] = random.uniform(0,1)
  in_h1 = np.zeros(k, dtype = float)
  temp = np.zeros(k, dtype = float)
  temp2 = np.zeros(k, dtype = float)
  in_h2 = np.zeros(k, dtype = float)
  out_h1 = np.zeros(k, dtype = float)
  out_h2 = np.zeros(k, dtype = float)
  in_o1 = np.zeros(k, dtype = float)
  out_o1 = np.zeros(k, dtype = float)
  partial_error_hidden1 = np.zeros(k, dtype = float)
  partial_error_hidden2 = np.zeros(k, dtype = float)
  partial_error_hidden3 = np.zeros(k, dtype = float)
  partial_error_hidden4 = np.zeros(k, dtype = float)
  partial_weights = np.zeros(6, dtype = float)
  sum_errors1 = np.zeros(k, dtype = float)
  sum_errors2 = np.zeros(k, dtype = float)

  N = 30000
  N_list = [*range(0, N, 1)]
  rmse_list = [*range(0, N, 1)]
  n = 0
  rmse = 1
  temp = np.arange(k,dtype = float)
  mse = np.arange(k,dtype = float)

  for n in range(N):
    in_h1 = weights[0] * inputs[:,0] + weights[2] * inputs[:,1]
    in_h2 = weights[1] * inputs[:,0] + weights[3] * inputs[:,1]
    temp = np.exp(-in_h1)
    temp2 = np.exp(-in_h2)
    out_h1 = (1/(1+temp))
    out_h2 = (1/(1+temp2))

    in_o1 = weights[4] * out_h1 + weights[5] * out_h2
    temp = np.exp(-in_o1)
    out_o1 = (1/(1+temp))
    sig_o1 = out_o1 * (1 - out_o1)
    sig_h1 = out_h1 * (1 - out_h1)
    sig_h2 = out_h2 * (1 - out_h2)
    error_o1 = (out_o1 - outputs)
    delta_o1 = error_o1 * sig_o1

    error_h1 = delta_o1 * weights[4]
    delta_h1 = error_h1 * sig_h1

    error_h2 = delta_o1 * weights[5]
    delta_h2 = error_h2 * sig_h2

    partial_weights[0] = inputs[:,0].T.dot(delta_h1)
    partial_weights[1] = inputs[:,0].T.dot(delta_h2)
    partial_weights[2] = inputs[:,1].T.dot(delta_h1)
    partial_weights[3] = inputs[:,1].T.dot(delta_h2)
    partial_weights[4] = out_h1.T.dot(delta_o1)
    partial_weights[5] = out_h2.T.dot(delta_o1)


    weights[0] = weights[0] - (L * partial_weights[0])
    weights[1] = weights[1] - (L * partial_weights[1])
    weights[2] = weights[2] - (L * partial_weights[2])
    weights[3] = weights[3] - (L * partial_weights[3])
    weights[4] = weights[4] - (L * partial_weights[4])
    weights[5] = weights[5] - (L * partial_weights[5])

    print("For Iteration: "+str(n))
    print("w1: "+str(weights[0]))
    print("w2: "+str(weights[1]))
    print("w3: "+str(weights[2]))
    print("w4: "+str(weights[3]))
    print("w5: "+str(weights[4]))
    print("w6: "+str(weights[5]))

    temp = np.power((out_o1 - outputs), 2)
    mse = temp.sum()/k
    rmse = math.sqrt(mse)
    print("RMSE is: "+ str(rmse))
    if(rmse < 0.15):
      break
    rmse_list[n] = rmse
    plt.plot(n, rmse,marker='o',color="blue", linewidth=1)

  f.write("Type of training: Batch\n"+"Learning Rate: "+str(L)+"Number of Iterations: "+str(n)+" RMSE: "+str(rmse)+"\n\n"+"Final weight1: "+str(weights[0])+"\nFinal weight2: "+str(weights[1])+"\nFinal weight3: "+str(weights[2])+"\nFinal weight4: "+str(weights[3])+"\nFinal weight5: "+str(weights[4])+"\nFinal weight6: "+str(weights[5])+"\n\nThe structure of the neural network is: 2 input nodes, 2 nodes in the hidden layer and 1 output node \nFollowed with two links to hidden node 1, one is from input1 with weight 1, the other is from input 2 with weight 3\nNode h2 has 2 links, 1 from input 1 with weight 2 and another from input 2 with weight 4\nThe final node has 1 link from h1 with weight 5 and 1 link from h2 with weight 6") # recording the file logs
  f.close()
  w1 = float(weights[0])
  w2 = float(weights[1])
  w3 = float(weights[2])
  w4 = float(weights[3])
  w5 = float(weights[4])
  w6 = float(weights[5])
  x1 = 0
  y= 0
  a = float(input("Enter the a variable: "))
  b = float(input("Enter the b variable: "))
  h1 = w1 * a + w3 * b
  h2 = w2 * a + w4 * b
  h1_out = (1/(1+math.exp(-h1)))
  h2_out = (1/(1+math.exp(-h2)))
  y = w5 * h1_out + w6 * h2_out
  y = (1/(1+math.exp(-y)))
  while(x1 == 0):

    if(y < 0.5):
      print("Predicted: "+str(y))
      print("The predicted output for a = "+ str(a)+" and b = "+str(b)+" is: 0")
    else:
      print("Predicted: "+str(y))
      print("The predicted output for a = "+ str(a)+" and b = "+str(b)+" is: 1")


    a = float(input("Enter the a variable or 5 to quit predicting: "))
    if(a == 5):
      break
    b = float(input("Enter the b variable or 5 to quit predicting: "))
    if(b == 5):
      break
    h1 = w1 * a + w3 * b
    h2 = w2 * a + w4 * b
    h1_out = (1/(1+math.exp(-h1)))
    h2_out = (1/(1+math.exp(-h2)))
    y = w5 * h1_out + w6 * h2_out
    y = (1/(1+math.exp(-y)))



  plt.xlabel('Iterations')
  plt.ylabel('RMSE Value')
  plt.title("RMSE progress and plot at each iteration N")
  plt.legend(["Iteration"], loc="upper right")
  plt.savefig('RMSE_plot.png')

elif(q == 2):
  inputs = np.array([[0, 0], [1, 0], [0, 1], [1, 1]])
  outputs = np.array([0,1,1,0], dtype = float)
  L = 0.1
  k = len(inputs)
  f = open("NNModelParameters.txt", "w")

  weights = np.random.randn(6,1)
  weights[0] = random.uniform(0,1)
  weights[1] = random.uniform(0,1)
  weights[2] = random.uniform(0,1)
  weights[3] = random.uniform(0,1)
  weights[4] = random.uniform(0,1)
  weights[5] = random.uniform(0,1)


  partial_error_hidden1 = np.zeros(k, dtype = float)
  partial_error_hidden2 = np.zeros(k, dtype = float)
  partial_error_hidden3 = np.zeros(k, dtype = float)
  partial_error_hidden4 = np.zeros(k, dtype = float)
  partial_weights = np.zeros(6, dtype = float)
  sum_errors1 = np.zeros(k, dtype = float)
  sum_errors2 = np.zeros(k, dtype = float)
  N = 30000
  n = 0
  rmse = 1
  temp = np.arange(k,dtype = float)
  mse = np.arange(k,dtype = float)

  for n in range(N):
    z = random.randint(0,3)
    in_h1 = weights[0] * inputs[z][0] + weights[2] * inputs[z][1]
    in_h2 = weights[1] * inputs[z][0] + weights[3] * inputs[z][1]
    out_h1 = (1/(1+math.exp(-in_h1)))
    out_h2 = (1/(1+math.exp(-in_h2)))

    in_o1 = weights[4] * out_h1 + weights[5] * out_h2
    out_o1 = (1/(1+math.exp(-in_o1)))
    sig_o1 = out_o1 * (1 - out_o1)
    sig_h1 = out_h1 * (1 - out_h1)
    sig_h2 = out_h2 * (1 - out_h2)
    error_o1 = (out_o1 - outputs[z])
    delta_o1 = error_o1 * sig_o1

    error_h1 = delta_o1 * weights[4]
    delta_h1 = error_h1 * sig_h1

    error_h2 = delta_o1 * weights[5]
    delta_h2 = error_h2 * sig_h2
    partial_weights[0] = inputs[z][0] * delta_h1
    partial_weights[1] = inputs[z][0] * delta_h2
    partial_weights[2] = inputs[z][1] * delta_h1
    partial_weights[3] = inputs[z][1] * delta_h2
    partial_weights[4] = out_h1 * delta_o1
    partial_weights[5] = out_h2 * delta_o1


    weights[0] = weights[0] - (L * partial_weights[0])
    weights[1] = weights[1] - (L * partial_weights[1])
    weights[2] = weights[2] - (L * partial_weights[2])
    weights[3] = weights[3] - (L * partial_weights[3])
    weights[4] = weights[4] - (L * partial_weights[4])
    weights[5] = weights[5] - (L * partial_weights[5])

    print("For Iteration: "+str(n))
    print("w1: "+str(weights[0]))
    print("w2: "+str(weights[1]))
    print("w3: "+str(weights[2]))
    print("w4: "+str(weights[3]))
    print("w5: "+str(weights[4]))
    print("w6: "+str(weights[5]))
    loss = out_o1 - outputs[z]
    print("Loss is: "+ str(loss))
    plt.plot(n, loss,marker='o',color="blue", linewidth=1)



  f.write("Type of training: Stochastic\n"+"Learning Rate: "+str(L)+"Number of Iterations: "+str(n)+" Loss: "+str(loss)+"\n\n"+"Final weight1: "+str(weights[0])+"\nFinal weight2: "+str(weights[1])+"\nFinal weight3: "+str(weights[2])+"\nFinal weight4: "+str(weights[3])+"\nFinal weight5: "+str(weights[4])+"\nFinal weight6: "+str(weights[5])+"\n\nThe structure of the neural network is: 2 input nodes, 2 nodes in the hidden layer and 1 output node \nFollowed with two links to hidden node 1, one is from input1 with weight 1, the other is from input 2 with weight 3\nNode h2 has 2 links, 1 from input 1 with weight 2 and another from input 2 with weight 4\nThe final node has 1 link from h1 with weight 5 and 1 link from h2 with weight 6") # recording the file logs
  f.close()
  w1 = float(weights[0])
  w2 = float(weights[1])
  w3 = float(weights[2])
  w4 = float(weights[3])
  w5 = float(weights[4])
  w6 = float(weights[5])
  x1 = 0
  y= 0
  a = float(input("Enter the a variable: "))
  b = float(input("Enter the b variable: "))
  h1 = w1 * a + w3 * b
  h2 = w2 * a + w4 * b
  h1_out = (1/(1+math.exp(-h1)))
  h2_out = (1/(1+math.exp(-h2)))
  y = w5 * h1_out + w6 * h2_out
  y = (1/(1+math.exp(-y)))

  while(x1 == 0):

    if(y < 0.5):
      print("Predicted: "+str(y))
      print("The predicted output for a = "+ str(a)+" and b = "+str(b)+" is: 0")
    else:
      print("Predicted: "+str(y))
      print("The predicted output for a = "+ str(a)+" and b = "+str(b)+" is: 1")

    a = float(input("Enter the a variable or press 5 to quit predicting: "))
    if(a == 5):
      break
    b = float(input("Enter the b variable or press 5 to quit predicting: "))
    if(b == 5):
      break
    h1 = w1 * a + w3 * b
    h2 = w2 * a + w4 * b
    h1_out = (1/(1+math.exp(-h1)))
    h2_out = (1/(1+math.exp(-h2)))
    y = w5 * h1_out + w6 * h2_out
    y = (1/(1+math.exp(-y)))

  plt.xlabel('Iterations')
  plt.ylabel('Predicted - Actual')
  plt.title("Loss and plot at each iteration N")
  plt.legend(["Iteration"], loc="upper right")
  plt.savefig('Loss_plot.png')
```
